```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/spectator_list.csv"
spectator_list_df = pd.read_csv(csv_file)
spectator_list_df.columns = ['Rank', 'number', 'title', 'points', 
              'price']
spectator_list_df.head()
```

```python
spectator_df = pd.DataFrame()
spectator_df['points'] =spectator_list_df['points']
spectator_df['price'] =spectator_list_df['price']
spectator_df['title'] =spectator_list_df['title']
spectator_df.head()
```

```python
wine_mag = "Resources/winemag-data-130k-v2.csv"
winemag_df = pd.read_csv(wine_mag)
winemag_df.head()
```

```python
cutwine_mag_df=winemag_df[:150]
```

```python
cutwine_mag_df
```

```python
# Create a filtered dataframe from specific columns
cutwine_magcol = ["country", "description", "designation","points","price","province","region_1","region_2","taster_name","taster_twitter_handle","title","variety","winery"]
newcutwine_magdf= cutwine_mag_df[cutwine_magcol].copy()

#Drop empty values 
newcutwine_magdf.dropna(subset=['price'], inplace=True)

#Delete columns
wine_mag_list2= newcutwine_magdf.drop(["country", "description", "designation","province","region_1","region_2","taster_name","taster_twitter_handle","variety","winery"], axis=1 )
```

```python
#Rename the column headers
wine_mag_list= wine_mag_list2.rename(columns={"Wine Name": "title",
                                                          "Wine Price": "price",
                                                          "Wine Points": "points"})
```

```python
wine_mag_list.dropna(inplace=True)
```

```python
merge_df = spectator_df.append(wine_mag_list, ignore_index=True)
merge_df
```

```python
#Create Database Connection
# <username>:<password>
connection_string = "postgres:postgres@localhost:5432/wine_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Load the Dataframes
wine_mag_list.to_sql(name='wine_mag', con=engine, if_exists='append', index=False)
merge_df.to_sql(name='wine_table', con=engine, if_exists='append', index=False)
spectator_df.to_sql(name='spectator_list', con=engine, if_exists='append', index=False)
```

```python
#Check that everything is good.
pd.read_sql_query('select * from wine_mag', con=engine).head()
```
